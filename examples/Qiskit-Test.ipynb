```python
# Import Libraries needed:
# Qiskit
# Matplotlib

import qiskit as q
import matplotlib.pyplot as plt 
```

```python
# Create QBits (2)
# Create Classical Bits (2)
# Create a Quantum Circuit with the classical and the QBits

qr = q.QuantumRegister(2) 
cr = q.ClassicalRegister(2)
circuit = q.QuantumCircuit(qr, cr) 
```

```python
# To perform Bell test, first we put in superposition the first QBit applying Hadamard Gate
circuit.h(qr[0]) 
circuit.draw(output='mpl')
```

```python
# We Include a CNot Gate to get both QBits in entanglement
circuit.cx(qr[0], qr[1]) 
circuit.draw(output='mpl')
```

```python
# Finally we add a measurement for both QBits
circuit.measure(qr, cr) 
circuit.draw(output='mpl')
```

```python
# Once we have our circuit ready, we run it on qasm_simulator in AER

simulator = q.Aer.get_backend('qasm_simulator') 
results = q.execute(circuit, backend=simulator).result()
q.visualization.plot_histogram(results.get_counts(circuit))
```
